```python
from sklearn.preprocessing import LabelEncoder
col1=['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status', 'Dependents']
col2=['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Dependents']
le= LabelEncoder()
for col in col1:
    df[col]=le.fit_transform(df[col])
for col in col2:
    test[col]=le.fit_transform(test[col])
```

```python
z=df.drop(columns=['Loan_Status'], axis=1)
x=z.values
y=df['Loan_Status'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
score= cross_val_score(classifier, x, y, cv=5)
print(np.mean(score)*100)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(x,y)
print(model.feature_importances_)
feat_importances = pd.Series(model.feature_importances_, index=z.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```
